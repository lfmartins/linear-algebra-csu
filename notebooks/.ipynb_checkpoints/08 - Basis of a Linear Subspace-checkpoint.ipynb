# Example 1, page 162

__Problem:__ Determine a basis for the subspace of $\mathbb{R}^4$ spanned by the vectors:

$$
\mathbf{u}_1=\left[\begin{matrix}2\\-1\\3\\1\end{matrix}\right]\quad
\mathbf{u}_2=\left[\begin{matrix}7\\-6\\5\\2\end{matrix}\right]
\mathbf{u}_3=\left[\begin{matrix}-3\\4\\1\\0\end{matrix}\right]
$$

This solution method uses the following result:

__Theorem.__ Suppose that two matrices $A$ and $B$ are equivalent. Then, the subspace generate by the rows of $A$ is equal to the subspace generated by the rows of $B$.

To use this theorem, we create a matrix that has the vectors $\mathbf{u}_1$, $\mathbf{u}_2$, $\mathbf{u}_3$ in its _rows_, and find its reduced row echelon form: 

```python
A = matrix_to_rational([[2, -1, 3, 1],[7,-6,5,2],[-3,4,1,0]])
A
```

```python
reduced_row_echelon(A)
```

The reduced row echelon form has only two linearly independent rows. The vectors contained in these rows generate the same subspace as the three vectors given in the problem. We conclude that a basis for the subspace is:

$$
\mathbf{v}_1=\left[\begin{matrix}1\\0\\13/5\\4/5\end{matrix}\right]\quad
\mathbf{v}_2=\left[\begin{matrix}0\\1\\11/5\\3/5\end{matrix}\right]
$$

A subspace has an infinite number of different bases. If we want a basis with integer values, we can multiply the vectors above by $5$:

$$
\mathbf{v}_1=\left[\begin{matrix}5\\0\\13\\4\end{matrix}\right]\quad
\mathbf{v}_2=\left[\begin{matrix}0\\5\\11\\3\end{matrix}\right]
$$

# Example 2, page 163

In this example, the same problem as in the previous example is solved by an alternative method. This method is more general, and is also used in the other examples in this section. It is based on the following result:

__Theorem.__ Let $A=[\mathbf{u}_1,\mathbf{u}_2,\ldots,\mathbf{u}_m]$ and 
$B=[\mathbf{v}_1,\mathbf{v}_2,\ldots,\mathbf{v}_m]$ be two equivalent matrices. Let $c_1,c_2,\ldots c_m$ be arbitrary scalars. Then:

$$
c_1\mathbf{u}_1+c_2\mathbf{u}_2+\ldots+c_m\mathbf{u}_m=c_1\mathbf{v}_1+c_2\mathbf{v}_2+\ldots+c_m\mathbf{v}_m=
$$

```python
A = matrix_to_rational([[2,7,-3],[-1,-6,4],[3,5,1],[1,2,0]])
A
```

```python
A1 = rop(A, 'R1*(1/2)=>R1', 'R1*(1)+R2=>R2', 'R1*(-3)+R3=>R3', 'R1*(-1)+R4=>R4')
A1
```

```python
A2 = rop(A1, 'R2*(-2/5)=>R2', 'R2*(-7/2)+R1=>R1', 'R2*(11/2)+R3=>R3', 'R2*(3/2)+R4=>R4')
A2
```

# Example 3, page 165

```python
A = matrix_to_rational([[1,3,-3,5],[2,6,8,-3],[5,15,19,-11],[-1,-3,-5,-2],[-4,-12,-18,-1]])
A
```

```python
A1 = rop(A, 'R1*(-2)+R2=>R2', 'R1*(-5)+R3=>R3', 'R1*(1)+R4=>R4', 'R1*(4)+R5=>R5')
A1
```

```python
A2 = rop(A1, 'R2*(1/14)=>R2', 'R2*(3)+R1=>R1', 'R2*(-34)+R3=>R3', 'R2*(8)+R4=>R4', 'R2*(30)+R5=>R5')
A2
```

```python
A3 = rop(A2, 'R3*(-7/31)=>R3', 'R3*(-31/14)+R1=>R1',  'R3*(13/14)+R2=>R2', 'R3*(31/7)+R4=>R4', 'R3*(62/7)+R5=>R5')
A3
```

# Example 4, page 165

```python
A = matrix_to_rational([[1,3,1,0,0],[1,2,0,1,0],[-2,-4,0,0,1]])
A
```

```python
A1 = rop(A, 'R1*(-1)+R2=>R2', 'R1*(2)+R3=>R3')
A1
```

```python
A2 = rop(A1, 'R2*(-1)=>R2', 'R2*(-3)+R1=>R1', 'R2*(-2)+R3=>R3')
A2
```

```python
A3 = rop(A2, 'R3*(1/2)=>R3', 'R3*(-3)+R1=>R1', 'R3*(1)+R2=>R2')
A3
```
